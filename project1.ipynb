```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\kodur\\Downloads\\archive.zip')
df.head(10)
```

```python
df.drop(columns=['id','Unnamed: 0'],inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
df['endDate'] = pd.to_datetime(df['endDate'], format='%b %d', errors='coerce')
df['startDate'] = pd.to_datetime(df['startDate'], errors='coerce')

# Replace endDate year with startDate year + 1
df['endDate'] = df.apply(
    lambda row: row['endDate'].replace(year=row['startDate'].year + 1),
    axis=1
)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import numpy as np
df['total_games']=df[['test','t20','odi']].sum(axis=1)
df['winrate']=np.round(df['matches']/df['squads'],2)
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\kodur\\Downloads\\archive.zip')
plt.figure(figsize=(20,9))

bars=plt.bar(df['name'], df['matches'],color='orange',label='Data')
plt.title('Analysis of Matches Played in Different Cricket Series',fontdict={'fontweight':'bold', 'fontsize':22})
plt.xlabel('Series Name',fontsize=22)
plt.ylabel('Number of Matches',fontsize=22)
plt.xticks(rotation=45, ha='right', fontsize=10, fontstyle='italic')
plt.legend()
plt.savefig('bar.png',dpi=300)
plt.show()

```

```python
###Line chart  ODI vs T20 trend
```

```python
plt.figure(figsize=(12,6),dpi=150)

plt.plot(df['odi'], df['t20'],
         color='orange',
         marker='^',
         linestyle= ':',
         markersize=10,
         linewidth=3,
         label='matches')
plt.title('Trends in ODI and T20 Matches',
          fontdict={'fontweight':'bold','fontsize':20},color='blue')
plt.xlabel('<------------------------------------ODI  Matches----------------------------->',
           fontsize=10,
          color='black')
plt.ylabel('<------------------------------------T20  Matches------------------------------>',fontsize=10
          ,color='black')
plt.yscale('log')

plt.xticks(ha='right', fontsize=10, fontstyle='italic')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
#Histogram  matches distribution:
```

```python
bins=[5,10,15,20,25,30]
plt.title('Distribution of Matches Played Across All Cricket Series',
          fontdict={'fontweight':'bold'}
         )
plt.hist(df['matches'],
         bins=bins,
         color='#F66BFA'
        )
plt.xlabel('Number of Matches')
plt.ylabel('Frequency')
plt.xticks(bins)
plt.tight_layout()
```

```python
###Pie chart  distribution of total games
```

```python
plt.style.use('default')
plt.figure(figsize=(14, 10))
plt.pie(df['total_games'],autopct='%1.1f%%'
        ,pctdistance=0.8,labeldistance=1.1)

plt.title("Distribution of total games",fontsize=18,fontweight='bold')
plt.legend(
    df['name'],
    loc="center left",
    bbox_to_anchor=(1.1, 0.5),
    fontsize=10
)

plt.show()
```
